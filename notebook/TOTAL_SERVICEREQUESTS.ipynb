```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data\SR2025.csv", encoding="latin1",on_bad_lines='skip')
```

```python
total_requests = len(df)
print(f"Total Requests: {total_requests}")
```

```python
plt.bar(["Total Requests"], [total_requests], color="skyblue")
plt.ylabel("Number of Requests")
plt.title("Total Service Requests")
plt.show()
```
