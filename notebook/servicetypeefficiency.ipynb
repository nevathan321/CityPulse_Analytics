```python
import pandas as pd

wards = {}

df = pd.read_csv("../data/SR2025.csv", encoding="latin1", on_bad_lines='skip')

for index, row in df.iterrows():
    ward = row["Service Request Type"]
    status = row["Status"]

    if ward not in wards:
        wards[ward] = {"Cases": 0, "Completed": 0}

    wards[ward]["Cases"] += 1
    if status == "Completed":
        wards[ward]["Completed"] += 1


most_efficient_ward = None
highest_efficiency = -1  

least_efficient_ward = None
lowest_efficiency = 2  

for ward, stats in wards.items():
    cases = stats["Cases"]
    completed = stats["Completed"]
    
    if cases > 0:
        efficiency = completed / cases
        stats["Efficiency"] = efficiency
        
    
        if efficiency > highest_efficiency:
            highest_efficiency = efficiency
            most_efficient_ward = ward
        
  
        if efficiency < lowest_efficiency:
            lowest_efficiency = efficiency
            least_efficient_ward = ward
    else:
        stats["Efficiency"] = None

print(f"Most efficient ward: {most_efficient_ward} with {highest_efficiency:.2%} efficiency")
print(f"Least efficient ward: {least_efficient_ward} with {lowest_efficiency:.2%} efficiency")
```

```python
import pandas as pd

wards = {}

df = pd.read_csv("../data/SR2025.csv", encoding="latin1", on_bad_lines='skip')

for index, row in df.iterrows():
    ward = row["Service Request Type"]

  
    if ward == "Business Complaint":
        continue

    status = row["Status"]

    if status == "Cancelled":
        continue

    if ward not in wards:
        wards[ward] = {"Completed": 0, "Incomplete": 0}

    if status in ["Completed", "Closed"]:
        wards[ward]["Completed"] += 1
    else:
        wards[ward]["Incomplete"] += 1

most_efficient_ward = None
highest_efficiency = -1

least_efficient_ward = None
lowest_efficiency = 2

for ward, stats in wards.items():
    total_cases = stats["Completed"] + stats["Incomplete"]

    if total_cases == 0:
        continue

    efficiency = stats["Completed"] / total_cases
    stats["Efficiency"] = efficiency

 
    if efficiency > highest_efficiency:
        highest_efficiency = efficiency
        most_efficient_ward = ward


    if stats["Incomplete"] > 0 and stats["Completed"] > 0 and efficiency < lowest_efficiency:
        lowest_efficiency = efficiency
        least_efficient_ward = ward

print(f"Most efficient ward: {most_efficient_ward} with {highest_efficiency:.2%} efficiency")
print(f"Least efficient ward: {least_efficient_ward} with {lowest_efficiency:.2%} efficiency")
```
